```r
df <- read.csv("measurements_16:24.csv",header=T)
library(ggplot2)
library(dplyr)
df %>% group_by(Size, Type) %>% summarise(Time = mean(Time)) -> df_mean
ggplot(df, aes(x = Size, y = Time, color = Type)) + geom_point(data=df) + geom_line(data=df_mean)

```
